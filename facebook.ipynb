```python
import pandas as pd
import numpy as np
import seaborn as sns
#import statsmodels.formula.api as smf

facebook = pd.read_csv('DATA/dataset_Facebook.csv', delimiter=";")
```

```python
facebook.info()
```

```python
facebook.describe()
```

```python
fb = facebook[facebook.share.notna()]
```

```python
fb = fb[fb.Paid.notna()]
```

```python
fb.info()
```

```python
fb.reset_index(inplace=True)
```

```python
fb.head()
```

```python
#Jackie's graph
sns.barplot(x='Type', y='Lifetime Post Consumers', data=fb, ci=None)
```

```python
groupedvalues = fb.groupby('Type').mean().reset_index()
groupedvalues.head()
```

```python
g =sns.barplot(x='Type',y='Lifetime Post Consumers',palette = 'Blues_d', data=groupedvalues)

for index, row in groupedvalues.iterrows():
    g.text(row.name,row['Lifetime Post Consumers'], round(row['Lifetime Post Consumers'],2), color='black', ha="center", size=12)
```

```python
#Jing's graph

grouped = fb.groupby('Post Month')
grouped = grouped.aggregate(np.mean).reset_index()
sns.set(rc={'figure.figsize': (10, 8)})
sns.set_style('ticks')
g = sns.barplot(x='Post Month',y='Lifetime Post Consumers',data=grouped,ci=None, color='royalblue')
for index, row in grouped.iterrows():
    g.text(row.name,row['Lifetime Post Consumers'], round(row['Lifetime Post Consumers'],2), color='black', ha='center',size=15)
```

```python
#MJ's graph

sns.set(rc={'figure.figsize': (14, 8)})
# sns.set_style('ticks')
sns.barplot(x = 'Post Weekday', y ='Lifetime Post Consumers', data = fb, hue = 'Paid', palette = 'Blues_d', ci = None)
```

```python
sns.set(rc={'figure.figsize': (14, 8)})
sns.barplot(x = 'Post Weekday', y ='Lifetime Post Consumers', data = fb, color = 'royalblue', ci = None)
```
